```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
E = 4
tau = 100e-6
```

```python
def impulse(t, E, tau):
    if abs(t)>tau/2:
        return 0.0
    return E

def spectrum(f, E, tau):
    if np.isclose(f, 0):
        return E*tau
    return E*tau*np.sinc(tau*f)

t_band = np.linspace(-2*tau, 2*tau, num=2**14)
f_band = np.linspace(-5/tau, 5/tau, num=2**14)
x = np.array([impulse(t, E, tau) for t in t_band])
Xf = np.array([spectrum(f, E, tau) for f in f_band])

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t_band*1e6, x)
plt.title("Импульс $x_1(t)$")
plt.xlabel("$t$, мкc")
plt.ylabel("$x_1(t)$, В")
plt.tight_layout()
plt.grid()

plt.subplot(2, 1, 2)
plt.title("АЧХ импульса $x_1(t)$")
plt.plot(f_band*1e-3, abs(Xf)*1e6)
plt.ylabel("$|X_1(f)|$, мкВ/Гц")
plt.xlabel("$f$, кГц")
plt.tight_layout()
plt.grid()

plt.savefig("A.png")
```

```python
def impulse(t, E, tau):
    if (t > tau) or (t < 0):
        return 0.0
    return E

def spectrum(f, E, tau):
    if np.isclose(f, 0):
        return E*tau
    return E*tau*np.sinc(f*tau) * np.exp(-1j*np.pi*f*tau)

t_band = np.linspace(-2*tau, 2*tau, num=2**14)
f_band = np.linspace(-5/tau, 5/tau, num=2**14)
x = np.array([impulse(t, E, tau) for t in t_band])
Xf = np.array([spectrum(f, E, tau) for f in f_band])

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t_band*1e6, x)
plt.title("Импульс $x_2(t)$")
plt.xlabel("$t$, мкc")
plt.ylabel("$x_2(t)$, В")
plt.tight_layout()
plt.grid()

plt.subplot(2, 1, 2)
plt.title("АЧХ импульса $x_2(t)$")
plt.plot(f_band*1e-3, abs(Xf)*1e6)
plt.ylabel("$|X_2(f)|$, мкВ/Гц")
plt.xlabel("$f$, кГц")
plt.tight_layout()
plt.grid()

plt.savefig("B.png")
```
